```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('C:\\Users\\shrey\\Desktop\\dm_office_sales.csv')

df.head()
```

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x='salary', y='sales', data=df, hue='level of education', palette='Dark2')
plt.show()
```

Seaborn - GRID PLOTS

```python
df.head()
```

```python
sns.catplot(data=df, x='gender', y='math score', kind='box', col='lunch', row='test preparation course')
```

```python
g=sns.PairGrid(df, hue='gender')

g = g.map_upper(sns.scatterplot)

g = g.map_diag(sns.histplot)

g = g.map_lower(sns.kdeplot)

g = g.add_legend()
```
