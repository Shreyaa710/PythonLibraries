```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('C:\\Users\\shrey\\Desktop\\dm_office_sales.csv')

df.head()
```

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x='salary', y='sales', data=df, hue='level of education', palette='Dark2')
plt.show()
```

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x='salary', y='sales', data=df, size='salary', alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x='salary', y='sales', data=df, s=200, style='level of education', hue='level of education')
plt.show()

plt.savefig("myplot.jpg")
```

```python
df = pd.read_csv('C:\\Users\\shrey\\Desktop\\dm_office_sales.csv')

df.head()
```

```python
#plt.figure(figsize=(5,8),dpi=200)
sns.rugplot(x='salary', data=df, height=0.5)
```

```python
sns.histplot(data=df, x='salary', kde=True)
```

```python
sns.kdeplot(data=df,x='salary')
```
