```python
import pandas as pd
```

Series Data Structure

```python
x = [3,4,5,6]

var = pd.Series(x,index=['a','s','d','f'],dtype='float',name="python")

print(var)
print(type(var))
print(var[2])
```

```python
dic = {"name":['python','c','c++','java'],"par":[12,13,14,15],"rank":[1,4,3,2]}

print(dic)
var1 = pd.Series(dic)
var2 = pd.DataFrame(dic)

print(var1)
print(var2)
```

```python
s = pd.Series(12, index=[1,2,3,4,5,6])

print(s)
print(type(s))
```

```python
s1 = pd.Series(12, index=[1,2,3,4,5,6])
s2 = pd.Series(12, index=[1,2,3,4])

print(s1+s2)
```

DataFrame Data Structure

```python
l = [1,2,3,4,5]

var = pd.DataFrame(l)

print(var)
print(type(var))
```

```python
d = {"a":[1,2,3,4,5],"s":[1,2,3,4,5], 1:[1,2,3,4,5]}

var = pd.DataFrame(d, columns=["a",1])

print(var)
print(type(var))
print(var["a"][3])
```

```python
list = [[1,2,3,4,5],[11,12,13,14,15]]

var = pd.DataFrame(list)

print(var)
```

```python
sr = {"s":pd.Series([1,2,3,4]), "r":pd.Series([1,2,3,4])}

var = pd.DataFrame(sr)

print(type(var))
print(var)
```

Arithmetic Operations

```python
var = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})

var
```

```python
var["C"] = var["A"]-var["B"]

var
```

```python
var["C"] = var["A"]*var["B"]

var
```

```python
var["C"] = var["A"]/var["B"]

var
```

```python
var = pd.DataFrame({"A":[10,20,30,40], "B":[15,16,17,18]})

var
```

```python
var["Python"] = var["A"] <= 20
var["Python_1"] = var["B"] >=16

var
```
