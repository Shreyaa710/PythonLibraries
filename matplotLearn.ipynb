Bar Plots

```python
plants = ['Spinach', 'Turnip', 'Rhubarb', 'Broccoli', 'Kale']
died = [10,25,5,30,21]
germinated = [74,88,56,69,59]
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants,germinated)
plt.title("GERMINATION RATES")
plt.ylabel("No. of seedings")
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants,died, color="red")
plt.title("Failure rates")
plt.ylabel("No. of seedings")
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants,died, color="red")
plt.bar(plants,germinated, width=0.4, color="blue", align="edge")
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants,died, color="red", bottom=germinated, label="failed seeding")
plt.bar(plants,germinated, color="blue", label="germinated seeding")
plt.legend(shadow=True, frameon=True, facecolor="white")
plt.title("Greenhouse seedings")
```

```python
plt.figure(figsize=(6,6))
plt.barh(plants,died, color="red")
```
